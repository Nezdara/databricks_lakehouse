## The Transformation Logic

```python
query = """
SELECT
    oi.order_id,
    oi.product_id,
    oi.seller_id,
    o.customer_id,
    o.purchase_timestamp,
    oi.price,
    oi.shipping_charges,
    (oi.price + oi.shipping_charges) AS total_item_cost
FROM
    silver.crm_order_items oi
JOIN
    silver.crm_orders o ON oi.order_id = o.order_id;
"""
df = spark.sql(query)
```

```python
df.limit(10).display()
```

## Writing Gold Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_fact_order_items")
```
