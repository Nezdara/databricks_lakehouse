## The Transformation Logic

```python
query = """
SELECT
    p.product_id,
    INITCAP(p.category_name) AS category_name, -- Clean up casing
    p.weight_g,
    p.length_cm,
    p.height_cm,
    p.width_cm,
    -- Calculate volume as a helpful business metric
    (p.length_cm * p.height_cm * p.width_cm) AS volume_cm3
FROM
    silver.crm_products p;
"""
df = spark.sql(query)
```

```python
df.limit(10).display()
```

## Writing Gold Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```
