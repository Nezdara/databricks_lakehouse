## Define Ingestion Configuration

```python
INGESTION_CONFIG = [
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/raw_sources/df_Customers.csv",
        "table": "df_Customers"
    },
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/raw_sources/df_OrderItems.csv",
        "table": "df_OrderItems"
    },
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/raw_sources/df_Orders.csv",
        "table": "df_Orders"
    },
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/raw_sources/df_Payments.csv",
        "table": "df_Payments"
    },
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/raw_sources/df_Products.csv",
        "table": "df_Products"
    }
]
```

## Ingest Files into Bronze Tables

```python
%python

for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} → workspace.bronze.{item['table']}")

    df = (
        spark.read
             .option("header", "true")
             .option("inferSchema", "true")
             .csv(item["path"])
    )

    (
        df.write
          .mode("overwrite")
          .format("delta")
          .saveAsTable(f"workspace.bronze.{item['table']}")
    )
```
